## Extract Function

Define the extract function that finds JSON file `bank_market_cap_1.json` and calls the function created above to extract data from them. Store the data in a `pandas` dataframe. Use the following list for the columns.


```python
columns=['Name','Market Cap (US$ Billion)']
```

```python
def extract_json(jsonfile):
    extracted_data = pd.DataFrame(columns=columns)
    extracted_data = extracted_data.append(extract_from_json(jsonfile), ignore_index=True)
    return extracted_data
    
def get_markets(filename):
    df_market = extract_json(filename)
    return df_market
    
get_markets("bank_market_cap_1.json")
```

<b>Question 1</b> Load the file <code>exchange_rates.csv</code> as a dataframe and find the exchange rate for British pounds with the symbol <code>GBP</code>, store it in the variable  <code>exchange_rate</code>, you will be asked for the number. Hint: set the parameter  <code>index_col</code> to 0.


```python
def extract_from_csv(file_to_process):
    dataframe = pd.read_csv(file_to_process)
    return dataframe

def find_currency(df_rates, search_currency = ''):
    df_rates.columns = ['currency','rate']

    index_col = 0
    while index_col < len(df_rates) and df_rates.iloc[index_col]['currency'] != search_currency:
        index_col += 1
    
    retDict = None
    if index_col < len(df_rates):
        retDict = {'position':index_col, 'currency': search_currency, 'rate': df_rates.iloc[index_col]['rate']}
    
    return retDict

def show_dict(dict):
    if dict != None:
        index_col     = dict['position']
        exchange_rate = dict['rate']
        currency_name = dict['currency']
        print("\nPosition:\t"+str(index_col)+"\nCurrency:\t"+currency_name+"\nRate:\t\t"+str(exchange_rate))
    else:
        print("\nCurrency '"+currency_name+"' not found in the table below:\n", df_rates)

def get_rates(filename):
    df_rates = extract_from_csv(filename)
    return(df_rates)

df_rates = get_rates("exchange_rates.csv")
print(df_rates)

uk_dict = find_currency(df_rates, 'GBP')
show_dict(uk_dict)

exchange_rate = uk_dict['rate']
exchange_rate
```

## Transform

Using <code>exchange_rate</code> and the `exchange_rates.csv` file find the exchange rate of USD to GBP. Write a transform function that

1.  Changes the `Market Cap (US$ Billion)` column from USD to GBP
2.  Rounds the Market Cap (US$ Billion)\` column to 3 decimal places
3.  Rename `Market Cap (US$ Billion)` to `Market Cap (GBP$ Billion)`


```python
market_filename = "bank_market_cap_1.json"
rates_filename = "exchange_rates.csv"

def extract(market_filename, rates_filename):
    df_market = get_markets(market_filename)
    df_rates = get_rates(rates_filename)

    return {'df_market': df_market, 'df_rate': df_rates}

def transform(extract_dict, from_currency, to_currency):
    transform_dict = extract_dict.copy()

    transform_dict['from'] = from_currency
    transform_dict['to'] = to_currency

    df_market = transform_dict['df_market'].copy() 
    df_rates = transform_dict['df_rate']

    from_dict = find_currency(df_rates, from_currency)
    to_dict = find_currency(df_rates, to_currency)
    transform_dict['from_dict'] = from_dict
    transform_dict['to_dict'] = to_dict

    from_rate = from_dict['rate']
    to_rate = to_dict['rate']
    factor = to_rate / from_rate

    actual_value = df_market['Market Cap (US$ Billion)']
    new_currency_title = 'Market Cap ('+to_currency+'$ Billion)'
    df_market.columns=['Name',new_currency_title]

    converted_value = actual_value * factor
    df_market[new_currency_title] = converted_value
    transform_dict['df_market'] = df_market

    return transform_dict
    
extract_dict = extract(market_filename, rates_filename)
#extract_dict

transform_dict = transform(extract_dict, 'USD', 'GBP')
transform_dict
```

## Load

Create a function that takes a dataframe and load it to a csv named `bank_market_cap_gbp.csv`. Make sure to set `index` to `False`.


```python
def load_market_csv(filename,df):
    df_load = df['df_market']
    if not df_load.empty:
        if filename != None and filename != "":
            df_load.to_csv(filename, index=False)
            
def load_market_json(filename,df):
    df_load = df['df_market']
    if not df_load.empty:
        if filename != None and filename != "":
            df_load.to_json(filename)

def load_csv(filename,df):
    if not df.empty:
        if filename != None and filename != "":
            df.to_csv(filename, index=False)
            
def load_json(filename,df):
    if not df.empty:
        if filename != None and filename != "":
            df.to_json(filename)

transform_dict = transform(extract_dict, 'USD', 'GBP')
transform_dict

load_market_csv('bank_market_cap_gbp.csv', transform_dict)
with open('bank_market_cap_gbp.csv', 'r') as f:
    lns = f.readlines()
            
load_market_json('bank_market_cap_gbp.json', transform_dict)
with open('bank_market_cap_gbp.json', 'r') as f:
    lns = f.readlines()
    
```

## Logging Function


Write the logging function <code>log</code> to log your data:


```python
def log(message, filename='log.log', mode='a'):
    timestamp_format = '%H:%M:%S %d/%h/%Y' 
    now = datetime.now() 
    timestamp = now.strftime(timestamp_format)
    with open(filename,mode) as f:
        f.write(timestamp + ',' + message + '\n')
```

## Running the ETL Process


Log the process accordingly using the following <code>"ETL Job Started"</code> and <code>"Extract phase Started"</code>


```python
# Write your code here
def ETL(market_filename, rates_filename, from_currency, to_currency, targetfile):
    # Log that you have started the ETL process
    logfilename = 'ETL.log'
    log("ETL Job Started: currency rates from '"+from_currency+"' to '"+to_currency+"'",logfilename,'w')

    # Log that you have started the Extract step
    log("Extract phase Started",logfilename)

    # Call the Extract function
    extracted_data = extract(market_filename, rates_filename)

    # Log that you have completed the Extract step
    log("Extract phase Ended",logfilename)

    # Log that you have started the Transform step
    log("Transform phase Started",logfilename)

    # Call the Transform function
    transformed_data = transform(extracted_data, from_currency, to_currency)

    # Log that you have completed the Transform step
    log("Transform phase Ended",logfilename)

    # Log that you have started the Load step
    log("Load phase Started",logfilename)

    # Call the Load function
    load_market_json(targetfile,transformed_data)

    # Log that you have completed the Load step
    log("Load phase Ended",logfilename)

    # Log that you have completed the ETL process
    log("ETL Job Ended",logfilename)

targetfile = 'bank_market_cap_gbp.json'
ETL(market_filename, rates_filename, 'USD','GBP', targetfile)
with open(targetfile, 'r') as f:
    lns = f.readlines()
    print(lns)
```
